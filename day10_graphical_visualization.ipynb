```python
a=np.arange(40)
b=np.random.randn(40,1)
```

```python
plt.plot(a,b)
```

```python
plt.plot(b)
```

```python
p=plt.figure(figsize=(12,6))       #
p=plt.plot(a,b) 
p=plt.title('Random Data Graph')
p=plt.grid()
p=plt.xlabel('x data-->')
p=plt.ylabel('y data-->')
p=plt.legend('Random Line')    #R on graph
p=plt.xticks(range(0,41,10),['Zero','Ten','Twenty','Thirty','Fourty'])    #intervals range
```

```python
#lw=line width , ls=line stle
p=plt.figure(figsize=(12,6))       #
p=plt.plot(a,b,color='y',marker='o',markerfacecolor='g',markeredgecolor='red',lw=3,ls='--',markersize=10) 
p=plt.title('Random Data Graph')
p=plt.grid()
p=plt.xlabel('x data-->')
p=plt.ylabel('y data-->')
p=plt.legend('Random Line')    #R on graph
p=plt.xticks(range(0,41,10),['Zero','Ten','Twenty','Thirty','Fourty'])  
p=plt.savefig('MyGraph.png',dpi=200)
```

```python
#Picture in picture graphs
fig=plt.figure(figsize=(10,6))
ax1= fig.add_axes([0,0,1,1]) #x,y,w,h
ax1.plot(a,b)
plt.ylim(-3,5)                #changing y axis limit
plt.xlim(-3,50)
plt.title('First Graph')
ax2= fig.add_axes([0.5,0.55,0.4,0.4])
ax2.hist(b)
plt.title('Second Graph')
```
