```python
import cv2
import skimage
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
img = cv2.imread('CRISTIANO-RONALDO.jpg') # BGR by default
plt.imshow(img)
```

```python
def imshow(im):
    cv2.imshow('image',im)
    cv2.waitKey(0)
```

```python
imshow(img)
```

### Color Space Conversion¶


```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imshow(img_gray)
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
imshow(img_hsv)
```

```python
cv2.COLOR_BGR2HSV
```

```python
cv2.COLOR_BGR2GRAY
```

```python
import skimage.color as skc
import skimage.io as skio
import skimage.transform as skt
```

```python
skio.imshow(img_rgb)
```

```python
img_gray2 = skc.rgb2gray(img_rgb)
skio.imshow(img_gray2)
```

```python
skio.imshow(img_gray)
```

### Cropping the Images


```python
img.shape
```

```python
plt.imshow(img)
```

```python
img_crop = img[100:100+350, 540:540+240, :].copy()
```

```python
imshow(img_crop)
```

```python
x,y,w,h = 540,100,240,350
```

```python
img_crop1 = img[y:y+h, x:x+w, :].copy()
imshow(img_crop1)
```

### Resizing Images


```python
img_intrp = cv2.resize(img, None, fx=0.5, fy=0.5, interpolation=cv2.INTER_CUBIC)
imshow(img_intrp)
```

```python
img_intrp.shape
```

```python
img_intrp = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
imshow(img_intrp)
print(img_intrp.shape)
```

```python
img_intrp2 = cv2.resize(img, (120,100), interpolation=cv2.INTER_CUBIC)
print(img_intrp2.shape)
imshow(img_intrp2)
```

### Image Rotation


```python
img_rot90 = cv2.warpAffine(img, cv2.getRotationMatrix2D((img.shape[1]/2,img.shape[0]/2),30,1), (img.shape[1],img.shape[0]))
imshow(img_rot90)
```

```python
img_rot_90 = skt.rotate(img,90,resize=True)
imshow(img_rot_90)
```

```python
img_rot_45 = skt.rotate(img,45,resize=True)
imshow(img_rot_45)
```
