```python
from minisom import MiniSom
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
img = plt.imread('elantra.png')
```

```python
def imshow(im):
    plt.imshow(im)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
imshow(img)
```

```python
img.shape
```

```python
img
```

```python
pixels = np.reshape(img, (img.shape[0]*img.shape[1],img.shape[2]))
pixels.shape
```

```python
som = MiniSom(x=3,y=3,input_len=3, sigma=0.1, learning_rate=0.2)
```

```python
som.random_weights_init(pixels)
```

```python
initial_weights = som.get_weights().copy()
initial_weights
```

```python
som.train_random(pixels,50,verbose=True)
```

```python
data_qt = som.quantization(pixels)
data_qt
```

```python
data_qt.shape
```

```python
k = data_qt.reshape(img.shape)
k.shape
```

```python
imshow(k)
```

```python
plt.imsave('car.png',k)
```
