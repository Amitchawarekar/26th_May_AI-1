```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import sklearn.datasets as skd
```

```python
data,labels = skd.make_blobs(n_samples = 400, n_features=2, centers=2, cluster_std = 4.1)
```

```python
data.shape
```

```python
sns.scatterplot(x=data[:,0], y=data[:,1],hue=labels)
```

```python

x = pd.DataFrame(data, columns=['A','B'])
x.head()
```

```python
labels
```

```python
x.describe()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest, ytrain,ytest = train_test_split(x,labels, test_size=0.25, random_state=1)
```

```python
model = KNeighborsClassifier(n_neighbors=4).fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
scores_train = []
scores_test = []
for i in range(2,50):
    model_i = KNeighborsClassifier(n_neighbors=i).fit(xtrain,ytrain)
    scores_train.append(model_i.score(xtrain,ytrain))
    scores_test.append(model_i.score(xtest,ytest))
```

```python
plt.figure(figsize=(15,6))
plt.plot(range(2,50),scores_train,marker='o')
plt.plot(range(2,50),scores_test,marker='*',color='red')
plt.grid(True)
plt.xticks(range(2,50))
0
```
