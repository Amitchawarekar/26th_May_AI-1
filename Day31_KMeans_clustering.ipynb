```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
import sklearn.datasets as skd
```

```python
data = skd.make_blobs(n_samples=500,n_features=2, centers=3, cluster_std=1.2,random_state=101,)
data
```

```python
df=pd.DataFrame(data[0])
y=pd.DataFrame(data[1])
df['y'] = y
```

```python
df.describe()
```

```python
df.groupby('y').describe()
```

```python
sns.scatterplot(df[0],df[1],c=df['y'])
```

```python
from sklearn.cluster import KMeans
```

```python
model= KMeans(n_clusters=3).fit(df.drop('y',axis=1))
```

```python
model.cluster_centers_
```

```python
model.labels_
```

```python
sns.scatterplot(df[0],df[1],c=model.labels_)
```

```python
sns.set_style('darkgrid')
```

```python
sns.set_palette('Set1')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.scatterplot(df[0],df[1],c=df['y'],cmap='Set1')
plt.title('Original')

plt.subplot(1,2,2)
sns.scatterplot(df[0],df[1],c=model.labels_,cmap='Set2')
plt.title('KMeans')
```

```python
model.inertia_
```

```python
model.inertia_
```

```python
model.score(df.drop('y',axis=1))
```

```python
inertia= []
for i in range(2,100):
    model_i = KMeans(n_clusters=i).fit(df.drop('y',axis=1))
    inertia.append(model_i.inertia_)
```

```python
plt.figure(figsize=(20,6))
plt.plot(range(2,100),inertia,lw=3,marker='o',markersize=8,ls='--',color='red',markeredgecolor='blue',markerfacecolor='blue')
plt.xticks(range(2,100))
plt.grid(True)
```
